Box plot:  https://en.wikipedia.org/wiki/Box_plot

```python
d1= randn(100)
d2 =randn(100)
```

```python
sns.boxplot([d1,d2])
```

```python
sns.boxplot([d1,d2],whis=np.inf)
```

```python
#normal dist
d1 = stats.norm(0,5).rvs(100)

#two gamma dist. Concatenated together
d2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                    -1* stats.gamma(5).rvs(50)])

#box plot
sns.boxplot([d1,d2],whis=np.inf)
```

```python
sns.violinplot([d1,d2])
```

```python
sns.violinplot(d2,bw=0.01)
```

```python
sns.violinplot(d1)
```

```python
sns.violinplot(d1,inner='stick')
```
