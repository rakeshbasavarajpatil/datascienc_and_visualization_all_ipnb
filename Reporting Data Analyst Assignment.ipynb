## Important Percentage Analysis part of the Growth Team

### 1145 Number of teams convert to paid within a week out of the 7735. As asked percentage of teams that convert to a paid subscription by week the trial started is about 14.80%

```python
sql_query = '''select count(teamid), (count(teamid)*1.0/7735 * 1.0) *100 as percentage_of_teams_doing_each_activity from dbtest
where highlights>0 and athletes>0 and coaches>0 and admins>0 and uploads>0 and viewers> 0 and days_to_convert <= 7 
'''
df_obs8 = pd.read_sql(sql_query,con)
df_obs8
```

### 34 teams do all the activities within a week of trial start to turning paid. So percentage of teams is 0.439% out of the total 7735 accounts created.

```python
sql_query = '''select count(teamid), (count(teamid)*1.0/7735 * 1.0) *100 as percentage_of_teams_doing_atleast_one_activity from dbtest
where (highlights>0 or athletes>0 or coaches>0 or admins>0 or uploads>0 or viewers> 0) and (days_to_convert <= 7) 
'''
df_obs9 = pd.read_sql(sql_query,con)
df_obs9
```

### 806 teams do at least one activity within a week of trial start to converting into paid. i.e. about 10.42% teams involved among the 7735 teams that created their account.


### It's very important that we increase the number of teams that engage in activities within the first week of trial start that will increase the chances of them getting converted to paid faster.
